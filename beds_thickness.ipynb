```python
# Import packages for use in notebook

import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
import matplotlib.colors as colors
import pandas as pd
from itertools import cycle, islice
import matplotlib as mpl
from scipy import stats
import seaborn as sns

%matplotlib inline
```

```python
pd.set_option('display.max_columns', 500) # Displays all columns in dataframe
df = pd.read_csv("..Data/az_beds_mm.csv") # Loads data from working directory and assigns name 'df'
```

```python
fig, axes = plt.subplots(ncols=3,nrows=1,figsize=(10,5)) # Plots figure and subplots

# Sets up color map

colors = ['#48a2ba','lightseagreen','mediumseagreen','seagreen',"gold","sandybrown"]
cmap = mpl.colors.ListedColormap(colors)
cmap.set_under("grey")
cmap.set_over("indianred")
norm = mpl.colors.LogNorm(vmin=1, vmax=100)

edgecolors = 'none'
linewidth=0.2
    
# Defines function for plotting bed thickness vs bed number

def graph_bed(x, y, x_label, y_label, title, titleweight, titleloc, ax):
        ax=ax
        ax.scatter(x, y, c=x ,s=25,cmap=cmap,norm=norm,zorder=12,
        edgecolors=edgecolors,linewidths=linewidth)
        ax.set_ylabel(y_label,weight='bold')
        ax.set_xlim(left=0,right=650)
        ax.set_ylim(bottom=0)
        ax.set_xlabel(x_label,weight='bold')
        ax.set_title(title,weight=titleweight,pad=10,loc=titleloc)
        ax.grid(zorder=0,linestyle='dashed')
        
# Plots subplots
       
graph_bed(df.beds_5C, df.Bed_Number,'Bed Thickness (cm)','Bed Number','Log **','bold','center',axes[0])
graph_bed(df.beds_4A, df.Bed_Number,'Bed Thickness (cm)','Bed Number','Log **','bold','center', axes[1])
graph_bed(df.beds_3A, df.Bed_Number,'Bed Thickness (cm)','Bed Number','Log **','bold','center', axes[2])

plt.suptitle('Beds', y=1.03, x=0.531, fontsize=14, weight='bold') # Plots title
plt.tight_layout()
```
